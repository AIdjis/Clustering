```python
from sklearn.mixture import GaussianMixture, BayesianGaussianMixture
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (silhouette_score, davies_bouldin_score, calinski_harabasz_score,)
```

```python
# Load data
dataset=pd.read_csv('data.csv')
dataset
```

```python
# plot the dataset
plt.scatter(dataset['X'],dataset['Y'])
plt.title('Data Set')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

##### Gaussian Mixture
* Description: Gaussian Mixture Clustering is a probabilistic model that represents data as a mixture of multiple Gaussian distributions. Each Gaussian component corresponds to a cluster, and the algorithm estimates the parameters (mean, covariance, and weight) of these distributions to fit the data. Unlike k-means, it can capture more complex cluster shapes by considering the variance within clusters. The Expectation-Maximization (EM) algorithm is typically used to optimize the model by iteratively assigning data points to clusters and updating the Gaussian parameters. This method is particularly useful for modeling data with overlapping clusters.

```python
gm = GaussianMixture(n_components=3, random_state=0)
cluster_labels=gm.fit_predict(dataset)
print("the best silhouette_score =", round(silhouette_score(dataset, cluster_labels),4))
print("the best davies_bouldin_score =", round(davies_bouldin_score(dataset, cluster_labels),4))
print("the best calinski_harabasz_score =", round(calinski_harabasz_score(dataset, cluster_labels),4))
```

```python
# plot the cluster 

plt.scatter(dataset['X'],dataset['Y'],c=cluster_labels)
plt.title('Gaussian Mixture')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

##### Bayesian Gaussian Mixture
* Description: The Bayesian Gaussian Mixture Clustering algorithm is a probabilistic model that extends the Gaussian Mixture Model (GMM) by incorporating a Bayesian framework. It assumes data is generated from a mixture of several Gaussian distributions, each representing a cluster. Unlike GMM, where the number of clusters is predefined, the Bayesian version uses a Dirichlet Process prior to allow the number of clusters to be automatically determined based on the data. This flexibility avoids overfitting by penalizing unnecessary clusters and provides more robust uncertainty estimates for the cluster assignments.

```python
BGM = BayesianGaussianMixture(n_components=3, random_state=0)
cluster_labels=BGM.fit_predict(dataset)
print("the best silhouette_score =", round(silhouette_score(dataset, cluster_labels),4))
print("the best davies_bouldin_score =", round(davies_bouldin_score(dataset, cluster_labels),4))
print("the best calinski_harabasz_score =", round(calinski_harabasz_score(dataset, cluster_labels),4))
```

```python
# plot the cluster 

plt.scatter(dataset['X'],dataset['Y'],c=cluster_labels)
plt.title('Bayesian Gaussian Mixture')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```
