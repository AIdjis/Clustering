```python
from sklearn.cluster import MeanShift
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (silhouette_score, davies_bouldin_score, calinski_harabasz_score,)
```

##### MeanShift
- The Mean Shift clustering algorithm :  is a non-parametric technique used for clustering data points in a feature space. It aims to locate the dense regions, or modes, of a dataset by iteratively shifting data points towards the nearest peak of the data distribution. The process begins with initializing the algorithm by placing a kernel (usually Gaussian) on each data point, which represents the region of influence. In each iteration, the algorithm computes the mean of the data points within the kernel's radius and then shifts the kernel towards this mean, effectively moving it towards regions of higher data point density. This shifting process continues until convergence, meaning the shifts become negligibly small, indicating the kernel has reached a mode.


```python
data=pd.read_csv('data.csv')
```

```python
#plot the dataset
plt.scatter(data['X'],data['Y'])
plt.title('Data Set')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

```python
# initialize and fit the MeanShift model

mean_shift = MeanShift(bin_seeding=True)  
cluster_labels = mean_shift.fit_predict(data)
print("the best silhouette_score =", round(silhouette_score(data, cluster_labels),4))
print("the best davies_bouldin_score =", round(davies_bouldin_score(data, cluster_labels),4))
print("the best calinski_harabasz_score =", round(calinski_harabasz_score(data, cluster_labels),4))
```

```python
plt.scatter(data['X'], data['Y'], c=cluster_labels)
plt.title('Mean Shift')
plt.show()
```
