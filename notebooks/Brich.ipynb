```python
from sklearn.cluster import Birch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (silhouette_score, davies_bouldin_score, calinski_harabasz_score,)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# load the dataset 
data=pd.read_csv('data.csv')
data
```

```python
# plot the dataset
plt.scatter(data['X'],data['Y'])
plt.title('Data Set')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

##### Birch

* Description: BIRCH (Balanced Iterative Reducing and Clustering using Hierarchies) is a hierarchical clustering algorithm designed for large datasets. It incrementally builds a tree structure, called a CF tree (Clustering Feature tree), that summarizes the dataset by storing key features of clusters in its nodes. BIRCH efficiently handles large datasets by building the tree in a single scan, and then refining it with additional scans if needed. It’s particularly effective for clustering data with noise and outliers, and its complexity makes it well-suited for handling massive datasets with limited memory.

```python
# initialize the and fit model 
# calculate the score of each evaluation metric
brc=Birch(n_clusters=3)
cluster_labels=brc.fit_predict(data)
print("the best silhouette_score = ", round(silhouette_score(data, cluster_labels),4))
print("the best davies_bouldin_score = ", round(davies_bouldin_score(data, cluster_labels),4))
print("the best calinski_harabasz_score = ", round(calinski_harabasz_score(data, cluster_labels),4))
```

```python
# plot the cluster 

plt.scatter(data['X'],data['Y'],c=cluster_labels)
plt.title('Birch Clustering')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```
